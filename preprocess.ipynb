```python
import cv2
import matplotlib.pyplot as plt
```

```python
def pre(path):
    img= cv2.imread(path) #Reads the image
  
    img= cv2.resize(img, (224, 224))    #resize the image 
    img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #Convert the image to grayscale
    img= img= img/255.0 #normalize the image.

    # It must be ensured that normalization takes place after using cv2.cvtColor() as cv2.cvtColor() expects pixel values between [0,255]
    # Normalization before converting color would throw an error
    return img

#Normalization results in pixel values between [0,1], but cv2 expects pixel values between [0,255]  
#So, we need to multiply the image by 255 before passing it to cv2.imshow() function. or use matplotlib to display it
```

```python
for i in range(3):
    path= f'image{i+1}.png'
    preprocessed_img= pre(path)
# While viewing the images, use either matplotlib to view it in grayscale using cmap='gray' or multiply the image by 255 before using cv2.imshow()
# Viewing the image using normal image viewers will give a complete black image, as it is a normalized image with values between [0,1]


    cv2.imwrite(f'preprocessed_{i+1}.png', preprocessed_img)
    plt.imshow(preprocessed_img, cmap='gray')
    plt.title(f'Preprocessed Image {i+1}')
    plt.show()
```
